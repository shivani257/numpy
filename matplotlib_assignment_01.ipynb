```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = ('https://raw.githubusercontent.com/hakimyameen/Datasets/main/CO2%20Emissions_Canada.csv')
path
```

```python
df = pd.read_csv(path)
df
```

```python
df.columns
```

```python
# univariate plot
plt.plot(df['Fuel Consumption City (L/100 km)'])
```

```python
plt.plot(df['CO2 Emissions(g/km)'])
```

```python
plt.plot(df['Engine Size(L)'])
```

```python
# groupby cylinders
data = df.groupby('Cylinders')['CO2 Emissions(g/km)'].mean()
data
```

```python
plt.plot(data,color='r',marker='o',linestyle='--')
plt.axis([0,21, 150,650]) # axis= ([xmin,xmax, ymin, ymax])
plt.title('CO2 Emission/Cyl')
plt.xlabel('CO2 Emissions(g/km)')
plt.ylabel('Cylinders')
plt.legend(['Co2 Emission/Cyl'],loc=4)  #loc:location
plt.show()
```

# ** Co2 Emission**

```python
plt.plot(data,color='r',marker='o',linestyle=':',linewidth=2,markersize=7,label='Co2 Emission/Cyl')
plt.title('Co2 Emission/Cylinder',fontsize=15)
plt.xlabel('Cylinders')
plt.ylabel('Co2 Emissions(g/km)')
plt.legend()
plt.grid(linestyle='--')
plt.show()
```

# **1. Fuel Consumption City (L/100 km)**

```python
data1 = df.groupby('Cylinders')['Fuel Consumption City (L/100 km)'].mean()
data1
```

```python
plt.plot(data1,label='Fuel Consumption City/Cyl',marker='*',markersize=7,linestyle=':',linewidth=2,color='r')
plt.xlabel('Cylinders')
plt.ylabel('Fuel Consumption City(L/100 km)')
plt.title('Fuel Consumption City/Cylinders')
plt.legend()
plt.grid(linestyle='--')
plt.show()
```

#2. **Fuel Consumption Hwy (L/100 km)**

```python
data2 = df.groupby('Cylinders')['Fuel Consumption Hwy (L/100 km)'].mean()
data2
```

```python
plt.plot(data2,label='Fuel Consumption Hwy (L/100 km)/Cyl',marker='>',markersize=7,linestyle=':',linewidth=2,color='y')
plt.xlabel('Cylinders')
plt.ylabel('Fuel Consumption Hwy(L/100 km)')
plt.title('Fuel Consumption Hwy/Cylinders')
plt.legend()
plt.grid(linestyle='--')
plt.show()
```

# **3.  Fuel Consumption Comb (L/100 km)**

```python
data3 = df.groupby('Cylinders')['Fuel Consumption Comb (L/100 km)'].mean()
data3
```

```python
plt.plot(data3,label='Fuel Consumption Comb(L/100 km)/Cyl',marker='o',markersize=7,linestyle=':',linewidth=2,color='g')
plt.xlabel('Cylinders')
plt.ylabel('Fuel Consumption Comb(L/100 km)')
plt.title('Fuel Consumption Comb/Cylinders')
plt.legend()
plt.grid(linestyle='--')
plt.show()
```

# **Combine 3 plots**

```python
plt.plot(data1,label='Fuel Consumption City/Cyl',marker='*',markersize=7,linestyle=':',linewidth=2,color='r')
plt.plot(data2,label='Fuel Consumption Hwy (L/100 km)/Cyl',marker='>',markersize=7,linestyle=':',linewidth=2,color='y')
plt.plot(data3,label='Fuel Consumption Comb(L/100 km)/Cyl',marker='o',markersize=7,linestyle=':',linewidth=2,color='g')
plt.xlabel('No of cylinders')
plt.ylabel('Fuel Consumptions')
plt.legend()
plt.grid(linestyle='--')
plt.show()
```
